```python
import csv
import matplotlib.pyplot as plt
import math

# Step 1: Read CSV file
marks = []
names = []

with open('marks.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  
    for row in reader:
        if len(row) >= 2: 
            names.append(row[0])
            try:
                marks.append(float(row[1]))
            except ValueError:
                print(f"Invalid mark value for {row[0]}: {row[1]}")

# Step 2: Calculate Mean
def calculate_mean(data):
    total = 0
    for num in data:
        total += num
    return total / len(data)

# Step 3: Calculate Variance
def calculate_variance(data, mean):
    total_diff_sq = 0
    for num in data:
        total_diff_sq += (num - mean) ** 2
    return total_diff_sq / len(data)

# Step 4: Calculate Standard Deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)

# Perform the calculations
mean = calculate_mean(marks)
variance = calculate_variance(marks, mean)
std_dev = calculate_std_dev(variance)

# Output the calculated values
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")

# Step 5: Write results back to CSV
with open('marks_with_stats.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Mark"])
    for name, mark in zip(names, marks):
        writer.writerow([name, mark])
    writer.writerow([])
    writer.writerow(["Mean", mean])
    writer.writerow(["Variance", variance])
    writer.writerow(["Standard Deviation", std_dev])

# Step 6: Plot
plt.plot(marks, 'bo-', label='Marks') 
plt.axhline(y=mean, color='g', linestyle='-', label=f'Mean: {mean}') # Mark mean
plt.axhline(y=mean + std_dev, color='r', linestyle='--', label=f'Mean + SD: {mean + std_dev}') # Mark standard deviation
plt.axhline(y=mean - std_dev, color='r', linestyle='--', label=f'Mean - SD: {mean - std_dev}') # Mark standard deviation
plt.title("Marks with Mean and Standard Deviation")
plt.xlabel("Index")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
def find_minimum_notes(amount):
    denominations = [2000, 500, 200, 100, 50, 20, 10, 5, 2, 1]
    note_count = {}
    for denom in denominations:
        if amount >= denom:
            note_count[denom] = amount // denom  #gives note counts
            amount = amount % denom  #sets remaining amount

    print("The minimum number of notes required are:")
    for note, count in note_count.items():
        print(f"{note}: {count} notes")

amount = 1339
find_minimum_notes(amount)
```

```python
# caesar_cipher encryption

def caesar_cipher(text, shift):
    encrypted_text = []
    for char in text:
        if char.isalpha():
            start = ord('a') if char.islower() else ord('A')
            encrypted_char = chr((ord(char) - start + shift) % 26 + start)
            encrypted_text.append(encrypted_char)
        else:
            encrypted_text.append(char)
    return ''.join(encrypted_text)
```

```python
# XORing with a salt

def xor_with_salt(text, salt):
    salt_repeated = (salt * (len(text) // len(salt) + 1))[:len(text)]
    return ''.join(chr(ord(c) ^ ord(s)) for c, s in zip(text, salt_repeated))
```

```python
#Encrypted file

def encrypt_file(filename, shift, salt):
    with open(filename, 'r') as file:
        content = file.read()
    
    # Apply Caesar cipher
    caesar_encrypted = caesar_cipher(content, shift)
    
    # Apply XOR with salt
    final_encrypted = xor_with_salt(caesar_encrypted, salt)
    
    # Write the encrypted content to a new file
    output_filename = filename + '_enc.txt'
    with open(output_filename, 'w') as file:
        file.write(final_encrypted)

    print(f"Encryption complete. Encrypted file saved as {output_filename}")

filename = 'input.txt' 
shift = 3  # Caesar cipher shift value
salt = 'abc'  # XOR salt

encrypt_file(filename, shift, salt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# from mymath import power, factorial


def power(base, exp):
    """Calculate base raised to the power of exp."""
    result = 1
    for _ in range(exp):
        result *= base
    return result

def factorial(n):
    """Calculate the factorial of n."""
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
def maclaurin_cos(x, terms=10):
    """Calculating the cosine of x using Maclaurin series."""
    cos_x = 0
    for n in range(terms):
        term = power(-1, n) * power(x, 2 * n) / factorial(2 * n)
        cos_x += term
    return cos_x

x_values = np.linspace(-2 * np.pi, 2 * np.pi, 400)
# print(x_values)
y_values = [maclaurin_cos(x, terms=10) for x in x_values]
# print(y_values)

# Calculate y values using numpy's cosine for comparison
y_actual = np.cos(x_values)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Maclaurin Series (10 terms)')
plt.plot(x_values, y_actual, label='Numpy Cosine', linestyle='--')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cosine Function: Maclaurin Series vs. Numpy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import json

def compare_dicts(dict1, dict2):
    differing_tags = {}
    
    # Iterate over keys in the 1st dictionary
    for key in dict1:
        if key in dict2:
            if isinstance(dict1[key], dict) and isinstance(dict2[key], dict):
                nested_diff = compare_dicts(dict1[key], dict2[key])
                if nested_diff:
                    differing_tags[key] = nested_diff
            elif dict1[key] != dict2[key]:
                differing_tags[key] = dict2[key] - dict1[key] if isinstance(dict1[key], (int, float)) and isinstance(dict2[key], (int, float)) else "Text change"
#             else:
#                 differing_tags[key] = "Only in 1"

                
        else:
            differing_tags[key] = "Only in 1"
    
    # Iterate over keys in the 2nd dictionary which is not in the 1st
    for key in dict2:
        if key not in dict1:
            differing_tags[key] = "Only in 2"
    
    return differing_tags


# JSON objects
json1 = {
    "x": 10.1,
    "y": 20,
    "name": "Anu",
    "dob": "2010-10-10"
}

json2 = {
    "x": 10,
    "y": 20,
    "name": "Ani",
    "dob": "2010-10-11",
    "z": 100
}

result = compare_dicts(json1, json2)
print(json.dumps(result, indent=4))
```
